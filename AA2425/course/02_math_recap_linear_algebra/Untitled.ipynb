```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Generate 3D Gaussian
mean = [2, -1, 3]
cov = [[2, 0.5, 0.2],
       [0.5, 1, 0.3],
       [0.2, 0.3, 1.5]]
num_points = 500

points = np.random.multivariate_normal(mean, cov, num_points)

# 2. Project onto unit sphere
norms = np.linalg.norm(points, axis=1, keepdims=True)
unit_points = points / norms

# 3. Compute centroids
original_centroid = points.mean(axis=0)
original_centroid_normalized = original_centroid / np.linalg.norm(original_centroid)
projected_centroid = unit_points.mean(axis=0)

# 4. Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot unit sphere (wireframe)
u, v = np.mgrid[0:2*np.pi:60j, 0:np.pi:30j]
xs = np.cos(u) * np.sin(v)
ys = np.sin(u) * np.sin(v)
zs = np.cos(v)
ax.plot_wireframe(xs, ys, zs, color='lightgray', alpha=0.3, linewidth=0.5)

# Scatter projected points
ax.scatter(unit_points[:, 0], unit_points[:, 1], unit_points[:, 2], color='royalblue', s=20, alpha=0.7, label='Projected Points')

# Plot centroids
ax.scatter(*original_centroid_normalized, color='green', s=150, edgecolor='black', label='Normalized Original Centroid')
ax.scatter(*projected_centroid, color='crimson', s=200, edgecolor='black', label='Centroid of Projected')

# Aesthetics
ax.set_title("3D Gaussian Projected onto Unit Sphere", fontsize=16)
ax.legend(fontsize=10)
ax.set_box_aspect([1,1,1])  # Equal aspect ratio
ax.grid(False)
#ax.off()
plt.axis('off')
plt.tight_layout()
plt.show()
```
